```python
import pandas as pd

def df_ordered(dataframe, column_name):
    """
    Ordena un DataFrame de mayor a menor según la columna especificada.

    :param dataframe: pd.DataFrame
        El DataFrame a ordenar.
    :param column_name: str
        El nombre de la columna por la que se desea ordenar.
    :return: pd.DataFrame
        El DataFrame ordenado de mayor a menor.
    """
    if column_name not in dataframe.columns:
        raise ValueError(f"La columna '{column_name}' no se encuentra en el DataFrame.")
    
    return dataframe.sort_values(by=column_name, ascending=False)
    
if __name__ == "__main__":
    data = {
        'Nombre': ['Alan', 'Bianca', 'Camila'],
        'Edad': [54, 19, 30],
        'Salario': [5698, 3849, 28739]
    }
    df = pd.DataFrame(data)

    df_ordenado = df_ordered(df, 'Edad')
    print(df_ordenado)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_bar(dataframe, column_name):
    """
    Crea un gráfico de barras para la columna especificada del DataFrame.

    :param dataframe: pd.DataFrame
        El DataFrame del cual se desea graficar una columna.
    :param column_name: str
        El nombre de la columna que se desea graficar.
    """
    if column_name not in dataframe.columns:
        raise ValueError(f"La columna '{column_name}' no se encuentra en el DataFrame.")
    
    value_counts = dataframe[column_name].value_counts()
    
 
    plt.figure(figsize=(10, 6))
    value_counts.plot(kind='bar', color='skyblue')
    plt.title(f'Gráfico por {column_name}')
    plt.xlabel(column_name)
    plt.ylabel('Frecuencia')
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()


if __name__ == "__main__":
  
    data = {
        'Nombre': ['Alan', 'Bianca', 'Camila', 'Alan', 'Bianca', 'Alan', 'Alan'],
        'Edad': [54, 19, 30, 54, 19, 54, 54],
        'Salario': [5698, 3849, 28739, 5698, 3849, 5698, 5698]
    }
    df = pd.DataFrame(data)
    plot_bar(df, 'Nombre')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

try:
    df = pd.read_csv('Mental.csv')
except FileNotFoundError:
    print("Error: El archivo 'Mental.csv' no fue encontrado.")
    exit()

columnas_interes = [
    "Entity",
    "Code",
    "Year",
    "Schizophrenia disorders (share of population) - Sex: Both - Age: Age-standardized",
    "Depressive disorders (share of population) - Sex: Both - Age: Age-standardized",
    "Anxiety disorders (share of population) - Sex: Both - Age: Age-standardized",
    "Bipolar disorders (share of population) - Sex: Both - Age: Age-standardized",
    "Eating disorders (share of population) - Sex: Both - Age: Age-standardized"
]

df_filtrado = df[columnas_interes]

df_filtrado.columns = [
    "Entity",
    "Code",
    "Year",
    "Schizophrenia",
    "Depression",
    "Anxiety",
    "Bipolar",
    "Eating"
]

df_agrupado = df_filtrado.groupby(['Entity', 'Code', 'Year']).mean().reset_index()

plt.figure(figsize=(15, 12))
features = ["Schizophrenia", "Depression", "Anxiety", "Bipolar", "Eating"]

for i, feature in enumerate(features):
    plt.subplot(3, 2, i + 1)
    sns.histplot(df_agrupado[feature], kde=True, bins=30, color='lightblue')
    plt.title(f'Distribución de {feature}')
    plt.xlabel(f'{feature} (%)')
    plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
try:
    df = pd.read_csv('Mental.csv')
except FileNotFoundError:
    print("Error: El archivo 'Mental.csv' no fue encontrado.")
    exit()

columnas_interes = [
    "Entity",
    "Code",
    "Year",
    "Schizophrenia disorders (share of population) - Sex: Both - Age: Age-standardized",
    "Depressive disorders (share of population) - Sex: Both - Age: Age-standardized",
    "Anxiety disorders (share of population) - Sex: Both - Age: Age-standardized",
    "Bipolar disorders (share of population) - Sex: Both - Age: Age-standardized",
    "Eating disorders (share of population) - Sex: Both - Age: Age-standardized"
]

df_filtrado = df[columnas_interes]

df_filtrado.columns = [
    "Entity",
    "Code",
    "Year",
    "Schizophrenia",
    "Depression",
    "Anxiety",
    "Bipolar",
    "Eating"
]

df_agrupado = df_filtrado.groupby(['Entity', 'Code', 'Year']).mean().reset_index()

correlation_matrix = df_agrupado[["Schizophrenia", "Depression", "Anxiety", "Bipolar", "Eating"]].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
plt.title('Correlación entre Trastornos Mentales')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

try:
    df = pd.read_csv('cancer_reg.csv')  # Asegúrate de que el archivo esté en el directorio correcto
except FileNotFoundError:
    print("Error: El archivo 'cancer_reg.csv' no fue encontrado.")
    exit()

x = df['avganncount']  # Por ejemplo, número promedio anual
y = df['target_deathrate']  # Tasa de mortalidad objetivo

slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

print(f'Pendiente (slope): {slope}')
print(f'Intersección (intercept): {intercept}')
print(f'R²: {r_value**2}')

plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='lightblue', label='Datos')
plt.plot(x, slope * x + intercept, color='orange', label='Línea de Regresión')
plt.xlabel('Promedio Anual de Casos')
plt.ylabel('Tasa de Mortalidad Objetivo')
plt.title('Regresión Lineal entre Avg Annual Count y Target Death Rate')
plt.legend()
plt.show()
```
