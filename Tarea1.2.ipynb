```python
import numpy as np
import matplotlib.pyplot as plt

freq = np.random.randint(1, 9, size=4)  

print("Frecuencias generadas:", freq)

amplitud = np.random.uniform(3, 6, size=4)

print("Amplitudes generadas:", amplitud)

t = np.arange(0, 1, 1/2000)

print("Arreglo t con np.arange:", t)

t = np.linspace(0, 1, num=2000)

freq = np.random.randint(1, 9, size=4)

amplitud = np.random.uniform(3, 6, size=4)

plt.figure(figsize=(10, 6))

for idx in range(len(freq)):
    y = amplitud[idx] * np.sin(2 * np.pi * freq[idx] * t)
    plt.plot(t, y, label=f'Onda {idx+1}: A={amplitud[idx]:.2f}, B={freq[idx]}')

plt.title('Ondas Sinusoidales')
plt.xlabel('Tiempo (t)')
plt.ylabel('Amplitud (y)')
plt.legend()
plt.grid()
plt.show()

t = np.linspace(0, 1, num=2000)

freq = np.random.randint(1, 9, size=4)

amplitud = np.random.uniform(3, 6, size=4)

y = np.zeros_like(t)

for idx in range(len(freq)):
    y += amplitud[idx] * np.sin(2 * np.pi * freq[idx] * t)

plt.figure(figsize=(10, 6))
plt.plot(t, y, label='Suma de Ondas Sinusoidales')
plt.title('Suma de Ondas Sinusoidales')
plt.xlabel('Tiempo (t)')
plt.ylabel('Amplitud (y)')
plt.legend()
plt.grid()
plt.show()

X = np.fft.fft(y)

frequencies = np.fft.fftfreq(len(X), d=t[1] - t[0])

plt.figure(figsize=(10, 6))
plt.plot(frequencies[:len(frequencies)//2], np.abs(X)[:len(X)//2])
plt.title('Transformada de Fourier de la Suma de Ondas')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.grid()
plt.show()

frequence = list(range(2000))

print("Frecuencia con range:", frequence)

X = np.fft.fft(y)

absX = np.abs(X)

print("Valor absoluto de X:", absX)

print("Primeros 10 elementos de absX:", absX[:10])

top_indices = np.argsort(absX[:10])[-4:]

top_values = absX[:10][top_indices]

print("Elementos mximos:", top_values)
print("ndices de los elementos mximos:", top_indices)

plt.stem(frequence[:10], absX[:10], 
         linefmt='b-', markerfmt='bo', basefmt='r-')
plt.title('Magnitud de la Transformada de Fourier (primeros 10 elementos)')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.grid()
plt.show()
```
