```python
import random
import numpy as np
import matplotlib.pyplot as plt

random.seed(5002)
numero_aleatorio = random.randint(-3, 3)
print("Núm aleatorio:", numero_aleatorio)
```

```python
numero_aleatorio_flotante = random.uniform(-10, 10)
print("Núm aleatorio flotante:", numero_aleatorio_flotante)
```

```python
ej1 = 0
ej2 = 1
numeros_normales = np.random.normal(loc=ej1, scale=ej2, size=1000)
print(numeros_normales[:10])
```

```python
plt.figure(figsize=(10, 6))
plt.hist(numeros_normales, bins=30, color='blue', edgecolor='pink', alpha=0.7)
plt.title('Distribución normal')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

```python
fig_1 = (ej1 - ej2, ej1 + ej2)
fig_2 = (ej1 - 2 * ej2, ej1 + 2 * ej2)
fig_3 = (ej1 - 3 * ej2, ej1 + 3 * ej2)

print(f"1σ: {fig_1}")
print(f"2σ: {fig_2}")
print(f"3σ: {fig_3}")
```

```python
rango_inferior = ej1 - ej2
rango_superior = ej1 + ej2
numeros_en_rango = np.sum((numeros_normales >= rango_inferior) & (numeros_normales <= rango_superior))
porcentaje = (numeros_en_rango / len(numeros_normales)) * 100
print(f"Porcentaje de números dentro del rango [-1σ, 1σ]: {porcentaje:.2f}%")
```

```python
rango_inferior = ej1 - 2 * ej2
rango_superior = ej1 + 2 * ej2
numeros_en_rango = np.sum((numeros_normales >= rango_inferior) & (numeros_normales <= rango_superior))
porcentaje = (numeros_en_rango / len(numeros_normales)) * 100
print(f"Porcentaje de números dentro del rango [-2σ, 2σ]: {porcentaje:.2f}%")
```

```python
rango_inferior = ej1 - 3 * ej2
rango_superior = ej1 + 3 * ej2
numeros_en_rango = np.sum((numeros_normales >= rango_inferior) & (numeros_normales <= rango_superior))
porcentaje = (numeros_en_rango / len(numeros_normales)) * 100
print(f"Porcentaje de números dentro del rango [-3σ, 3σ]: {porcentaje:.2f}%")
```

```python
def analizar_distribucion(mu, sigma, total):
    numeros_normales = np.random.normal(loc=ej1, scale=ej2, size=total)
    plt.figure(figsize=(10, 6))
    plt.hist(numeros_normales, bins=30, color='red', edgecolor='green', alpha=0.7)
    plt.title(f'Histograma de {total} números con distribución normal')
    plt.xlabel('Valor')
    plt.ylabel('Frecuencia')
    plt.show()
    
    ej2_1_inferior = ej1 - ej2
    ej2_1_superior = ej1 + ej2
    ej2_2_inferior = ej1 - 2 * ej2
    ej2_2_superior = ej1 + 2 * ej2
    ej2_3_inferior = ej1 - 3 * ej2
    ej2_3_superior = ej1 + 3 * ej2
    
    numeros_1sigma = np.sum((numeros_normales >= ej2_1_inferior) & (numeros_normales <= ej2_1_superior))
    numeros_2sigma = np.sum((numeros_normales >= ej2_2_inferior) & (numeros_normales <= ej2_2_superior))
    numeros_3sigma = np.sum((numeros_normales >= ej2_3_inferior) & (numeros_normales <= ej2_3_superior))

    porcentaje_1sigma = (numeros_1sigma / total) * 100
    porcentaje_2sigma = (numeros_2sigma / total) * 100
    porcentaje_3sigma = (numeros_3sigma / total) * 100
    
    print(f"\nPorcentaje de números dentro de los rangos:")
    print(f"  [-1σ, 1σ]: {porcentaje_1sigma:.2f}%")
    print(f"  [-2σ, 2σ]: {porcentaje_2sigma:.2f}%")
    print(f"  [-3σ, 3σ]: {porcentaje_3sigma:.2f}%")

ej1 = 0
ej2 = 1.0
total = 1000

analizar_distribucion(ej1, ej2, total)
```

```python
analizar_distribucion(0, 1.0, 1000)
analizar_distribucion(3, 5.0, 5000)
analizar_distribucion(-3, 3.0, 10000)
```
